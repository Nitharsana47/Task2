```python
import pandas as pd
df=pd.read_csv("Titanic-Dataset.csv")
print("summary statistics")
print(df.describe(include="all"))
```

```python
print("\nMean values:\n", df.mean(numeric_only=True))
```

```python
print("\nMedian values:\n",df.median(numeric_only=True))
```

```python
print("\n Standard deviation:\n",df.std(numeric_only=True))
```

```python
print("\n mode :\n",df.mode().iloc[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=df['Age'])
plt.title('Age')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=df['Fare'])
plt.title('Fare')
plt.show()
```

```python
plt.hist(df['Age'].dropna(),bins=30,color='blue',edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Fare'].dropna(),bins=30,color='salmon',edgecolor='black')
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

for col in numeric_cols:
    # Histogram
    plt.figure()
    plt.hist(df[col], bins=20, color='lightblue')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

    # Boxplot
    plt.figure()
    sns.boxplot(x=df[col], color='orange')
    plt.title(f"Boxplot of {col}")
    plt.show()
```
